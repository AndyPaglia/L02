```python
# Importing functions and libraries that will be used in the code.
import os
import torch
import json
import torchaudio
from lpmc.music_captioning.model.bart import BartCaptionModel
from lpmc.utils.audio_utils import load_audio, STR_CH_FIRST
from lpmc.utils.eval_utils import load_pretrained
from lpmc.music_captioning.model.modules import AudioEncoder
from omegaconf import OmegaConf
```

```python
# Code that allows for more flexibility for configuration.
# In this code we do not use this part since the parameter are already been defined after. This would give more flexibility in the case
# the user wants to change and give manually the parameters.

import argparse

parser = argparse.ArgumentParser(description='PyTorch MSD Training')
parser.add_argument('--gpu', default=0, type=int,
                    help='GPU id to use.')
parser.add_argument("--framework", default="transfer", type=str)
parser.add_argument("--caption_type", default="lp_music_caps", type=str)
parser.add_argument("--max_length", default=128, type=int)
parser.add_argument("--num_beams", default=5, type=int)
parser.add_argument("--model_type", default="last", type=str)
parser.add_argument("--audio_path", default="", type=str)
```

```python
# Directory of the folder containing the audio files.
audio_path = '/content/drive/MyDrive/L02CapstoneProject/FakeMusicCaps'
```

```python
import numpy as np

# Function for loading and preprocessing the audio files.
def get_audio(audio_path, duration=10, target_sr=16000):
    n_samples = int(duration * target_sr)
    audio, sr = load_audio(
        path=audio_path,
        ch_format=STR_CH_FIRST,
        sample_rate=target_sr,
        downmix_to_mono=True,
    )
    if len(audio.shape) == 2:
        audio = audio.mean(0, False)

    # Do padding if necessary.
    input_size = int(n_samples)
    if audio.shape[-1] < input_size:
        pad = np.zeros(input_size)
        pad[:audio.shape[-1]] = audio
        audio = pad

    # Suddivision in 10 second long audio clips. All the folders contain 10 second audio clips. The only one with longer files is SunoCaps.
    ceil = int(audio.shape[-1] // n_samples)
    audio = torch.from_numpy(np.stack(np.split(audio[:ceil * n_samples], ceil)).astype('float32'))
    return audio

# Function to take every file from the folder (there are subfolders to be checked).
def load_audio_files_from_folder(folder_path):
    audio_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".wav"):
                audio_files.append(os.path.join(root, file))
    return audio_files

audio_folder = '/content/drive/MyDrive/L02CapstoneProject/FakeMusicCaps'
audio_files = load_audio_files_from_folder(audio_folder)
```

```python
import os
import json

def save_captions_to_drive(captions, audio_file, output_root_dir):

    # Extract the subfolder from FakeMusicCaps.
    relative_folder = os.path.relpath(os.path.dirname(audio_file), '/content/drive/MyDrive/L02CapstoneProject/FakeMusicCaps')

    # Extract the name of the audio file.
    file_name = os.path.basename(audio_file)

    # Create the correlated folder in the output folder.
    output_folder = os.path.join(output_root_dir, relative_folder)
    os.makedirs(output_folder, exist_ok=True)

    # Create the directory.
    captions_file = os.path.join(output_folder, file_name.replace(".wav", "_captions.json"))

    # Save captions as JSON files.
    with open(captions_file, 'w') as f:
        json.dump(captions, f, indent=4)
```

```python
# Here we define the parameter that were defined in the "argparse" function, but here are defined with a specific value.
# We define all the parameters directly in the code.

class Args:
    gpu = 0
    framework = "transfer"
    caption_type = "lp_music_caps"
    max_length = 128
    num_beams = 5
    model_type = "last"
    audio_path = ""

args = Args()
```

```python
# Captioning function: produces captions from the wav files in input.
def captioning(args, audio_files, output_root_dir):

    save_dir = "/content/drive/MyDrive/L02CapstoneProject/lp-music-caps/lp-music-caps/exp/transfer/lp_music_caps"
    config = OmegaConf.load(os.path.join(save_dir, "hparams.yaml"))

    # Loading of the model.
    model = BartCaptionModel(max_length=config.max_length)
    model, save_epoch = load_pretrained(args, save_dir, model, mdp=config.multiprocessing_distributed)

    torch.cuda.set_device(args.gpu)
    model = model.cuda(args.gpu)
    model.eval()
    i = 0

    # Loop to create the captions.
    for audio_path in audio_files:
        i = i+1
        if i % 100 == 0:
          print(f"Captioning number {i} : {audio_path}")

        # Load the audio file.
        audio_tensor = get_audio(audio_path)
        if args.gpu is not None:
            audio_tensor = audio_tensor.cuda(args.gpu, non_blocking=True)

        # Caption generation.
        with torch.no_grad():
            output = model.generate(samples=audio_tensor, num_beams=args.num_beams)

        # Print the result.
        inference = {}
        number_of_chunks = range(audio_tensor.shape[0])
        for chunk, text in zip(number_of_chunks, output):
            time = f"{chunk * 10}:00-{(chunk + 1) * 10}:00"
            item = {"text": text, "time": time}
            inference[chunk] = item

        # Saves the captions in the directory of reference for the output.
        save_captions_to_drive(inference, audio_path, output_root_dir)


output_root_dir = "/content/drive/MyDrive/L02CapstoneProject/FakeMusicCapsCAPTIONS2500"
captioning(args, audio_files, output_root_dir)
```
