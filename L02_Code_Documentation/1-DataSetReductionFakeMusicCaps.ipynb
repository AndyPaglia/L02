**DATASET REDUCTION**

Code that allows to reduce the folder FakeMusicCaps from 33056 to 15069.

This reduction of the dataset is a must process when using Google Colab since Google Colab has a limited free time for using the GPU and for the runtime.

```python
import os
import random
import shutil

# Mounting Google Drive in Google drive to allow using certain data folders useful for the project.
from google.colab import drive
drive.mount('/content/drive')

# Directories to find the input folder to reduce and the output folder to put the results.
source_root = "/content/drive/MyDrive/L02CapstoneProject/FakeMusicCaps"
dest_root = "/content/drive/MyDrive/L02CapstoneProject/FakeMusicCaps2500"

# Number of elements to mantain for each folder.
max_files_per_class = 2500

# Create destination root.
os.makedirs(dest_root, exist_ok=True)

# Loop for each subfolder of the folder.
for class_folder in os.listdir(source_root):
    class_path = os.path.join(source_root, class_folder)
    if not os.path.isdir(class_path):
        continue

    # Creates an output folder with the same configuration of the input one.
    dest_class_path = os.path.join(dest_root, class_folder)
    os.makedirs(dest_class_path, exist_ok=True)

    all_files = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
    random.shuffle(all_files)
    selected_files = all_files[:max_files_per_class]

    for f in selected_files:
        src_file = os.path.join(class_path, f)
        dest_file = os.path.join(dest_class_path, f)
        shutil.copy2(src_file, dest_file)

print(f" Reduction Completed: {max_files_per_class} files for each subfolder.")
```
